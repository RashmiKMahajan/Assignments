Q2. W.A.P to accept cost-price of Bike and display the road tax to be paid according to following Criteria:
    TAX       Cost
    15%     >100000
    10%     >50000 and <=100000
    5%       <= 50000
    

```python
cost = int(input("Enter the Cost-price of the Bike :"))

if cost>100000 :
    Tax = 0.15*cost
    print("Road Tax to be paid is : ", Tax)
elif cost>50000 and cost <=100000 :
    Tax = 0.10*cost
    print("Road Tax to be paid is : ", Tax)
else :
    Tax = 0.05*cost
    print("Road Tax to be paid is : ", Tax)
```

Q5. Why and when to use While loop in Python. Give datailed Description with Example?

Python while loop is used to run a specific code until a certain condition is met. If the condition evaluates to True, the code inside the while loop is executed. condition is evaluated again.

The while loop in Python is used to iterate over a block of code as long as the test expression (condition) is true. We generally use this loop when we don't know the number of times to iterate beforehand. In the while loop, test expression is checked first. The body of the loop is entered only if the test_expression evaluates to True.

```python

# Python program to illustrate
# while loop
count = 0
while (count < 3):
    count = count + 1
    print("pwskills Data-science-Masters")
```

```python
# In the above example, the condition for while will be True as long as the counter variable (count) is less than 3.  counter variable (count) is less than 3. 
```

Q6.  Use nested while loop to display 3 different patterns

```python
#Pattern 1
count =0
i = 1
while  i <= 5 :
    j = 1
    while j <= i :
        print(count,end = '')
        count += 1
        j += 1
    print("")    
    i += 1
```

```python
#Pattern 2
i = 1
while  i <= 5 :
    j = 1
    while j <= i :
        print('*',end = '')
        j += 1
    print("")    
    i += 1
```

```python
#Pattern 3
i = 5
while  i >= 0 :
    j = 0
    while j <= i :
        print('*',end = '')
        j += 1
    print("")    
    i -= 1
```
