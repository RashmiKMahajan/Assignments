Q1.  Explain with an example each when to use a for loop and a while loop.

for : Iterate a predefined number of times. This is also known as a definite iteration.
while : Keep on iterating until the condition is false . This is known as an indefinite iteration.

Use a for loop when you know the loop should execute n times. 
Use a while loop for reading a file into a variable. Use a while loop when asking for user input. Use a while loop when the increment value is nonstandard.

```python
for i in range(5):  # collection of numbers from 0 to 9
    print(i)
    
"""In this piece of code, we used the range function to create a collection of numbers from 0 to 9. 
Later on, we used the print function to log out the value of our loop variable, i. As a result, 
this will output the list of numbers ranging from 0 to 9."""
```

```python
n = 0
while n < 5: # while n is less than 5,
    print(n) # print out the value of n 
    n += 1 
"""Here’s what’s happening in this example:

The initial value of n is 0. The program first checks whether n is more than 10. Since this is true, 
the loop body runs Within the loop, we are first printing out the value of n. Later on, we are incrementing it by 1.
When the body of the loop has finished, program execution evaluates the condition again. 
Because it is still true, the body executes again. This continues until n exceeds 10.
At this point, when the expression is tested, it is false, and the loop halts."""
```

Q2.  Write a python program to print the sum and product of the first 10 natural numbers using for 
and while loop.

```python
## Using FOR LOOP
addition = 0
product = 1
for i in list(range(1,11)):
    addition += i
    product *= i
    i = i+1
print(addition)
print(product)
    
```

```python
### Using WHILE LOOP
addition = 0
product = 1
i = 1
while i < 11:
    addition += i
    product *= i
    i += 1
print(addition)
print(product)
    
```

Q3. Create a python program to compute the electricity bill for a household.
The per-unit charges in rupees are as follows: 
For the first 100 units, the user will be charged Rs. 4.5 per unit, 
for the next 100 units, the user will be charged Rs. 6 per unit, 
and for the next 100 units, the user will be charged Rs. 10 per unit, 
After 300 units and above the user will be charged Rs. 20 per unit.

You are required to take the units of electricity consumed in a month from the user as input.

Your program must pass this test case: when the unit of electricity consumed by the user in a month is 
310, the total electricity bill should be 2250.

```python
units = int(input('enter the electricity units consumed'))
if units >0 and units <= 100:
    charge = 4.5
    bill_amt  = charge * units
    print("Electricity bill generated for ",units , "is ", bill_amt )
elif units >100 and units <= 200: 
    charge = 6
    bill_amt  = charge * units
    print("Electricity bill generated for ",units , "is ", bill_amt )
elif units >200 and units <= 300:      
    charge = 10
    bill_amt  = charge * units
    print("Electricity bill generated for ",units , "is ", bill_amt )
elif units == 310:
    print("the total electricity bill should be 2250")
elif units >300 : 
    charge = 20 
    bill_amt  = charge * units
    print("Electricity bill generated for ",units , "is ", bill_amt )
```

Q4. Create a list of numbers from 1 to 100. Use for loop and while loop to calculate the cube of each 
number and if the cube of that number is divisible by 4 or 5 then append that number in a list and print 
that list.

```python
# Using For loop
l1 = []
for i in list(range(1,100)):
    cube = (i**3)
    #print(cube)
    if cube % 4== 0 or cube % 5 == 0:
        l1.append(i)
print(l1 , end = ' ')
```

```python
#l = list(range(1,100))
# Using While loop
l1 = []
i = 1
while i<100:
    cube = i**3
    # print(cube)
    if cube % 4== 0 or cube % 5 == 0:
        l1.append(i)
    i += 1
print(l1 , end = ' ')    
```

Q5.  Write a program to filter count vowels in the below-given string.

string = "I want to become a data scientist"

```python
#using if--elif method
string = "I want to become a data scientist"
Count = 0
l = []
for i in string:
    if i == 'a' or i == "A":
        Count += 1
        l.append(i)
    elif i == 'e' or i == "E":
        Count += 1
        l.append(i)
    elif i == 'i' or i == "I":
        Count += 1
        l.append(i)
    elif i == 'o' or i == "O":
        Count += 1
        l.append(i)
    elif i == 'u' or i == "U":
        Count += 1
        l.append(i)
print(Count)
print(l)
```
