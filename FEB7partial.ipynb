```python
 #Sort a list of tuples having fruit names and their quantity. [("mango",99),("orange",80), ("grapes", 1000)-
fruits = [("mango",99),("orange",80), ("grapes", 1000)]
fruits
```

```python
# Display Before Sorting 
for i in fruits:
    print(i)
```

```python
# sorting by quqntity
f = fruits.sort(key = lambda x : x[1])
```

```python
# After sorting by quantity
for i in fruits:
    print(i)
```

```python
# sorting by fruit names (a to z)
f = fruits.sort(key = lambda x : x[0])
```

```python
# After sorting by fruits name
for i in fruits:
    print(i)
```
